```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv('iris.csv')
```

```python
data.describe()
```

```python
# Summary statistics by species
a=data.groupby('variety')
b=a.get_group('Setosa')
c=a.get_group('Virginica')
d=a.get_group('Versicolor')

print("Setosa\n",b.describe())
print("\nVirginica\n",c.describe())
print("\nVersicolor\n",d.describe())
```

```python
## Histogram
sb.histplot(data=data['sepal.length'],kde=True)
plt.title('Distribution of sepal.length')
plt. xlabel='sepal.length'
plt.ylabel='Count'
plt.show()

sb.histplot(data=data['sepal.width'],kde=True)
plt.title('Distribution of sepal.length')
plt. xlabel='sepal.width'
plt.ylabel='Count'
plt.show()

sb.histplot(data=data['petal.length'],kde=True)
plt.title('Distribution of petal.length')
plt. xlabel='sepal.length'
plt.ylabel='Count'
plt.show()

sb.histplot(data=data['petal.width'],kde=True)
plt.title('Distribution of sepal.length')
plt. xlabel='petal.width'
plt.ylabel='Count'
plt.show()
```

```python

# Pair plot of the dataset
sb.pairplot(data, hue='variety', height=2.5)
plt.show()
```

```python
# Box plot for each feature by species
for column in data.columns[:-1]:
    plt.figure(figsize=(10, 6))
    sb.boxplot(x='variety', y=column, data=data)
    plt.title(f'Box plot of {column} by species')
    
```

```python
df=data.copy()
df['variety']=df['variety'].replace({'Setosa':0,'Versicolor':1,'Virginica':2})
```

```python
sb.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# Violin plot for each feature by species
for column in data.columns[:-1]:
    plt.figure(figsize=(10, 6))
    sb.violinplot(x='variety', y=column, data=data, inner='quartile')
    plt.title(f'Violin plot of {column} by species')
    
```

```python
# Strip plot for each feature by species
for column in data.columns[:-1]:
    plt.figure(figsize=(10, 6))
    sb.stripplot(x='variety', y=column, data=data,jitter=True)
    plt.title(f'Strip plot of {column} by species')
```

```python
# Joint plot for two features (example: sepal_length and sepal_width)
sb.jointplot(x='sepal.length', y='sepal.width', data=data, hue='variety', kind='scatter', height=8)
plt.show()
```
